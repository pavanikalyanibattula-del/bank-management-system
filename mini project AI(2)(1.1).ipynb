```python
def display_stops():
    print("""Stops Available:
1. Central Bus Stand
2. Market Junction
3. City Hospital
4. Railway Station
5. University Gate
6. Tech Park
7. Airport
""")
```

```python
stops = {
    1: "Central Bus Stand",
    2: "Market Junction",
    3: "City Hospital",
    4: "Railway Station",
    5: "University Gate",
    6: "Tech Park",
    7: "Airport"
}
```

```python
stop_distance = {
   1:5,
   2:9,
   3:14,
   4:18,
   5:23,
   6:29,
   7:35,
    
}
```

```python
def get_passenger_details():
    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    boarding = int(input("Enter Boarding Stop number: "))
    destination = int(input("Enter Destination Stop number: "))
    return name, age, boarding, destination
```

```python
def calculate_distance(boarding, destination):
    if boarding < destination:
        return stop_distance[destination] - stop_distance[boarding]
    return 0
```

```python
def calculate_fare(distance, age):
    if age < 18:
        return distance
    elif age >= 60:
        return distance * 2 * 0.6
    return distance * 2
    
```

```python
def print_ticket(name, age, boarding, destination):
    distance = calculate_distance(boarding, destination)
    fare = calculate_fare(distance, age)
    print("\n------ BUS TICKET ------")
    print("Passenger Name:", name)
    print("Passenger Age:", age)
    print("Boarding Stop:", stops[boarding])
    print("Destination:", stops[destination])
    print("Distance:", distance, "KM")
    print("Final Fare: ₹", fare)
```

```python
display_stops()
name, age, boarding, destination = get_passenger_details()
print_ticket(name, age, boarding, destination)
```
